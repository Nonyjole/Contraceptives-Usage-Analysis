# 1: What are the main reasons for using or not using contraceptives among women in Lagos Nigeria?

# 3: What is the overall percentage of women who use condoms in Lagos Nigeria?

```python
# First filter those who uses Contraceptives 

Only_Contraceptives = Contraceptives_Dataset[Contraceptives_Dataset['Use Contraceptive'] == 'Yes']
Only_Contraceptives
```

```python
# Then do a value_counts of the observations of those who use Contraceptives


Contraceptives_Usage_By_Numbers  = Only_Contraceptives['Type of Contraceptive'].value_counts()
Contraceptives_Usage_By_Numbers
```

```python
# Draw a pie chart of the value counts to see the percentage of the population uses condoms 


Percentage_Of_Women_Using_Condoms  = Only_Contraceptives['Type of Contraceptive'].value_counts()

The_Colors = np.array(['green','cyan', 'blue', 'black', 'pink', 'yellow', 'purple', 'grey', 'teal', 'orange', 'red',  'gold', 'lime'])
plt.pie(Percentage_Of_Women_Using_Condoms, autopct = '%1.1f%%', colors = The_Colors)
plt.legend(Percentage_Of_Women_Using_Condoms.index, bbox_to_anchor = (2,0.99), loc = 'upper right')
plt.title('The Percentage of women who use condom in Lagos')
plt.show()
```

## From the chart above we can see that the Percentage of women who use condom in Lagos is about 35.2%. Or, it also means that 35.2% of women who use contraceptives prefer condoms.

# 4: How does the usage of condoms vary by age, marital status, number of kids, and LGA?

```python
# Filter condoms users from the contraceptives users and Check the distribution of usage by age

Condoms_Uasage = Only_Contraceptives[Only_Contraceptives['Type of Contraceptive'] == 'Condom']
Condoms_Uasage 
```

```python
# Check how usage of condoms varies by age

By_Age = Condoms_Uasage['Age'].value_counts()

By_Age.plot(kind = 'barh')

plt.title('THE USAGE OF CONDOMS  BY AGE')
```

## From the chart above, it means that it is women who are 25 years old that uses condoms the most.Whereas women aged 24, 18, 26, 27, 33, 35, 22, 31 and 40 that uses condoms the ieast as a birth control measure.

```python
# Check how usage of condoms varies by marital status

By_Marital_Status = Condoms_Uasage['Marital Status'].value_counts()

My_Explode = [0.2,0]
plt.pie(By_Marital_Status, labels  = By_Marital_Status, explode = My_Explode)
plt.legend(By_Marital_Status.index,  bbox_to_anchor = (2,0.5), loc = 'upper right')
plt.title('THE USAGE OF CONDOMS  BY MARITAL STATUS')
plt.show()
```

## This means that majority of women who use condoms and has no kids are single(unmarried) women.

## The usage of condoms by marital status shows that single sexually active women uses condoms the most as their birth control methods.


```python
# Check how usage of condoms varies by number of kids

Number_Of_Kids = Condoms_Uasage['Number of Kids'].value_counts()

plt.pie(Number_Of_Kids, labels  = Number_Of_Kids)
plt.legend(Number_Of_Kids.index, loc = 'upper right',  bbox_to_anchor = (2,0.5))
plt.title('THE USAGE OF CONDOMS BY NUMBER OF KIDS')
plt.show()
```

## This means that 16 users has no kids, 2 users has 3 kids and 1 user has 2 kids

```python
# Check how usage of condoms varies by LGAs

By_Age = Condoms_Uasage['LGA'].value_counts()

By_Age.plot(kind = 'bar')

plt.title('THE USAGE OF CONDOMS  BY LGA')
```

## This shows that most condoms users as their type of contraceptives lives in Alimosho
